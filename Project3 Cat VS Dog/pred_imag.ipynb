In this file, we will use the model we trained to predict a image.

```python
import torch
import matplotlib.pyplot as plt
from main import Model, DEVICE, transform
from PIL import Image
```

```python
# Set class names
class_names: list[str] = ['cat', 'dog']
```

```python
# Import model
pth_path: str = 'cat_vs_dog.pth'
model = Model(load_pth=True, pth_path=pth_path)
model.cat_vs_dog.eval()
```

```python
# Import image
image_PIL = Image.open('imag_2000/test/dog/dog.1034.jpg')
image_tensor: torch.Tensor = transform(image_PIL)
image_tensor = image_tensor.unsqueeze(0).to(DEVICE)
```

```python
# Plot this image
plt.imshow(image_PIL)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
# Pridict
out = model(image_tensor)
pred = out.argmax(dim=1).item()
print(class_names[pred])
```

```python
# Another example
image_PIL = Image.open('./imag_2000/test/cat/cat.1034.jpg')
image_tensor: torch.Tensor = transform(image_PIL)
image_tensor = image_tensor.unsqueeze(0).to(DEVICE)

# Pridict
out = model(image_tensor)
pred = out.argmax(dim=1).item()
print(class_names[pred])

# Plot this image
plt.imshow(image_PIL)
plt.xticks([]), plt.yticks([])
plt.show()
```
